## Use Case - Pearson Korrelationskoeffizient
Wie Sie bereits in den Folien gesehen haben lassen sich lineare Zusammenhänge zwischen mehreren Variablen durch den Pearson Korrelationskoeffizienten darstellen. Diese Erkenntnisse können verwendet werden um bisher unbekannte Muster oder etwaige Kausalzusammenhänge aufzudecken. Solche können dann für das Feature Engineering sehr hilfreich sein. Versuchen Sie doch einmal selbst Korrelationen zwischen verschiedenen Variablen aufzudecken !

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
## Laden Sie das mpg-Dataset
mpg = pd.read_csv("../Datasets/mpg.csv")
```

Wie Sie in der Formel ableiten können, lässt sich die Korrelation nach Pearson durch die Kovarianz und die Standardabweichung ausrechnen. 

Ihre Aufgabe ist es nun eine Funktion zu programmieren,welche Ihnen den Pearson-Korrelationskoeffizienten für 2 übergebene Features ausrechnet.

Die Funktionen die Sie dabei benutzen können sind beispielsweise np.cov(), df.cov() [df = DataFrame] für die Kovarianz und df.std() für die Standardabweichung.

```python
def calculate_corr(feature1, feature2):
    corr = 0
    cov_x_y = np.cov(feature1,feature2)[0][1]
    std_x = feature1.std()
    std_y = feature2.std()
    corr = cov_x_y / (std_x*std_y)
    return corr
```

Sie können Ihre Lösung mit folgendem Code überprüfen

```python
print(calculate_corr(mpg.displacement, mpg.cylinders))
calculated_corr = mpg[['displacement','cylinders']].corr().reset_index().displacement[1]
print(f"richtige Korrelation: {calculated_corr}")
```

Schauen Sie sich die gerade berechnete Korrelation noch einmal in einem Pairplot an um sich über die Art der Beziehung zwischen den beiden Variablen näher bewusst zu werden.

```python
sns.pairplot(mpg[['cylinders','displacement','origin']], hue='origin', palette={1:'green',2:'red',3:'blue'})
```
